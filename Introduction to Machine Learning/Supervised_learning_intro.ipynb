## What is Supervised Learning?

üëâ In **Supervised Learning**, the algorithm learns from data that already has the correct answer (called **labels or targets**).

## üßæ Think of it like:

A teacher giving questions along with answers to help a student learn how to solve similar problems later.

## üéì Real-Life Examples:

| Example                        | Input (Features)                | Output (Target/Label) | Type           |
| ------------------------------ | ------------------------------- | --------------------- | -------------- |
| Predict house price            | Size, location, no. of bedrooms | House price ‚Çπ         | Regression     |
| Will a customer buy insurance? | Age, income, job                | Yes / No              | Classification |
| Email spam detection           | Email content (text)            | Spam / Not Spam       | Classification |

## Types of Supervised Learning
    
Supervised learning has two major types:

**1. üîµ Regression**

Goal: Predict a number

Example: Predict salary, temperature, price, marks, etc.

Example algorithms: Linear Regression, Decision Tree Regression

**2. üü¢ Classification**

Goal: Predict a category

Example: Yes/No, Male/Female, Fraud/Not Fraud, Disease/No Disease

Example algorithms: Logistic Regression, KNN, Decision Tree Classifier

## Important Terminology

| Term                | Meaning                                         |
| ------------------- | ----------------------------------------------- |
| üéØ **Target/Label** | What we want to predict (e.g., price, category) |
| üìä **Feature**      | The inputs used to predict (e.g., age, size)    |
| üß† **Model**        | The algorithm that learns from the data         |
| üõ†Ô∏è **Train**       | Teaching the model using data                   |
| üîç **Test**         | Checking how well the model predicts new data   |

## Mini Practical Example ‚Äî Predict House Price üè†

We‚Äôll use a super basic example:

Predict house price based on its size using Linear Regression.

**üß∞ 1. Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# üìå What does each library do?
# pandas For working with data tables (like Excel)
# numpy For numerical operations
# matplotlib.pyplot For plotting charts
# sklearn.linear_model Contains the LinearRegression model
```

**üßÆ 2. Create Simple Data**

```python
x = np.array([500, 750, 1000, 1250, 1500]).reshape(-1,1) # X: These are input (independent) values. Input feature: House size in square feet
#.reshape(-1, 1)
# This reshapes the array from 1D to 2D because many ML models (like scikit-learn's Linear Regression) require 2D input.
# -1 means: ‚Äúautomatically figure out the number of rows‚Äù
# 1 means: ‚Äú1 column‚Äù 

y = np.array([30, 45, 60, 75, 90]) # y: These are output (dependent) values. # This creates a 1D array of the target/output values: # Output/label: Price in ‚Çπ (in thousands)
```

**üß† 3. Train the Model**

```python
model = LinearRegression()
model.fit(x,y)
# LinearRegression() Creates the model
# model.fit(X, y) Trains the model on input X and output y

#Now the model has learned the relationship between size and price.
```

**üìà 4. Visualize the Line the Model Learned**

```python
plt.scatter(x, y, color="blue", label="Actual") # label='Actual': Adds a label so that we know these are actual data points (will be shown in the legend). To visualize the real data, These are the original values you gave to the model.
plt.plot(x, model.predict(x), color="red", label="Predicted Line")
# plt.plot: Creates a line plot.
# model.predict(x): Predicts the prices for the same x values using your trained model.
# color='red': Line will be red.
# label='Predicted Line': This label helps show it's the line your model learned.

plt.xlabel("House Size (sq ft)")
plt.ylabel("House price (Rs. in Thousands)")
plt.title("Simple Linear Regression")
plt.legend() # Displays the labels you gave earlier: Blue = Actual,Red = Predicted Line
plt.show()

# plt.scatter only shows real data.
# You won‚Äôt see how well the model fits or predicts unless you draw the red line using plt.plot.
# You need both:
# Blue dots to show what really happened.
# Red line to show what the model thinks should happen.
# In short:
# plt.scatter	Shows actual data points
# plt.plot	Shows the model's prediction line
# Other lines	Add labels, titles, and show the plot
```

**üîç 5. Predict for a New House**

```python
new_house_size = np.array([[1100]]) # we give it in shape (1, 1) ‚Üí 1 row, 1 column ‚Üí like this: [[1100]] because, Machine learning models expect 2D arrays for input features, not a plain number or 1D array.

predicted_price = model.predict(new_house_size)
# This uses the trained linear regression model to make a prediction.
# It outputs an array : predicted_price = [66.0]
# (for example, if the model learned: price = 0.06 * size ‚Üí 0.06 √ó 1100 = 66)
# So, your model predicts that a house of 1100 sq ft should cost ‚Çπ66,000 (in thousands).

print(f"Predicted Price: Rs.{predicted_price[0]*1000:.2f}")
# predicted_price[0]: This gets the first (and only) predicted value from the array.
# *1000: You multiply it by 1000 to convert from ‚Äúin thousands‚Äù to actual ‚Çπ.
# Because your original y values were like [30, 45, 60, ...] which meant ‚Çπ30,000, ‚Çπ45,000, etc.
# :.2f: This formats the number to 2 decimal places.
```
